```python
# pmdarima is a library with models to handle time series predictive modeling
from pmdarima import auto_arima

# SARIMAX is the seasonal autoregressive integrated moving average for handling time series forcasting
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.model_selection import train_test_split
```

```python
# Create a datetime index for each month in each year
# pd.melt is a way of taking the columns as a categorical value and changing it into row data
retailmelt = pd.melt(retail, id_vars=['YEAR'], value_vars=['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC'], var_name='MONTH', value_name='VALUE')

# Create a date from the year and month
retailmelt['DATE'] = pd.to_datetime(retailmelt['YEAR'].astype(str) + retailmelt['MONTH'], format='%Y%b')

# Set the date as the index and sort by it
retailmelt = retailmelt.set_index('DATE').sort_index()

retailmelt
```

```python
retailmelt = retailmelt.drop(columns=['YEAR','MONTH'])
```

```python
# to get a train test split, set a date to split on, the df has been ordered with a date index, so the < or > will work as a filter

splitdate = '2020-07-01'
train = retailmelt.loc[retailmelt.index < splitdate]
test = retailmelt.loc[retailmelt.index >= splitdate]
```

```python
test = test.dropna()
test
```

```python
# stepwise fit is a way of going through the data automatically to assign the best parameters

retailstep = auto_arima(train['VALUE'], seasonal=True, m=12, trace=True, suppress_warnings=True, stepwise=True)
```

```python
# order will set the order to be called in while fitting the model

order = retailstep.order
seasonal_order = retailstep.seasonal_order
```

```python
# set the params of the model
model = SARIMAX(train, order=order, seasonal_order=seasonal_order)
sarima = model.fit()
```

```python
forecast = sarima.forecast(steps=12)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
rmse = np.sqrt(mean_squared_error(test['VALUE'], forecast))
rmse
```

The RMSE is around 50k which is not terrible considering the values are in the hundreds of thousands. Its a larger margin of error than I would prefer, but it solid for this work
